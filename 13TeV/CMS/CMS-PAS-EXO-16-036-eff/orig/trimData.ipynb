```python
import numpy as np
import glob
```

```python
for f in glob.glob('./*_cons_*.dat'):
    with open(f,'r') as fdata:
        header = fdata.readlines()[0]
    print(f)
    header = header.replace('#','').split()
    mrec = [eval(x.replace('mrec','')) for x in header if 'mrec' in x][0]
    ihscp = header.index('mhscp')
    data = np.loadtxt(f,dtype={'names' : header, 'formats' : [float]*len(header)})
    trim_data = data[data['mhscp']>=mrec/0.6]
    trim_data.sort(order='mhscp')
    fnew = f.replace('.dat','_trim.txt')
    np.savetxt(fnew,trim_data,delimiter='  ',header = "        ".join(header),fmt='%.6e')
    print(fnew)
```
